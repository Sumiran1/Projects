```python
from urllib.request import urlopen
from bs4 import BeautifulSoup
import pandas as pd
import numpy as np
```

Scraping rookie year stats from basketball refrence, since we have compiled a set of data from 1985-2019, we will have to fidn the query names from this list and isolate those players.

```python
def get_rookie_stats(year):
    url= "https://www.basketball-reference.com/awards/awards_{}.html#all_roy".format(year)
    html = urlopen(url)
    soup = BeautifulSoup(html)
    
    headers = [th.getText() for th in soup.findAll('tr',limit=2)[1].findAll('th')]
    headers = headers[1:]
    
    rows = soup.findAll('tr')[1:]
    player_stats = [[td.getText() for td in rows[i].findAll('td')] for i in range(len(rows))]
    stats = pd.DataFrame(player_stats, columns = headers)
    return stats.drop_duplicates().dropna(how='all')
```

```python
get_rookie_stats(2019)
```

```python
def filter_mvps(year):
    rookie_mvp= get_rookie_stats(year)
    pts_max_array= rookie_mvp['Pts Max'].unique()
    rookies_no_mvps= rookie_mvp[rookie_mvp['Pts Max'] == pts_max_array[1]]
    return rookies_no_mvps
```
